# **1. Business Understanding**

### Background:
WirelessUS menghadapi tantangan dalam mempertahankan pelanggan di tengah persaingan ketat industri telekomunikasi. Tingginya angka *customer churn* menunjukkan perlunya strategi retensi yang lebih baik. Saat ini, perusahaan menggunakan metode rule-based sederhana untuk mengidentifikasi pelanggan yang berisiko churn, namun metode ini kurang efektif dalam mendeteksi pola churn yang kompleks. Oleh karena itu, proyek ini bertujuan untuk membangun model prediksi churn berbasis *machine learning* yang lebih akurat guna mengidentifikasi pelanggan yang berisiko berhenti berlangganan, sehingga perusahaan dapat mengambil tindakan preventif yang lebih efektif.

### Problem Statement:
- Bagaimana WirelessUS dapat memprediksi pelanggan yang akan churn secara lebih akurat dibandingkan dengan metode rule-based yang saat ini digunakan?
- Bagaimana model prediksi baru ini dapat membantu WirelessUS mengidentifikasi faktor-faktor utama yang mempengaruhi churn dan meningkatkan strategi retensi pelanggan?

### Goals:
1. Mengembangkan model prediksi churn berbasis *machine learning* yang lebih baik dari metode rule-based sebelumnya.
2. Mengidentifikasi variabel kunci yang mempengaruhi churn pelanggan untuk mendukung pengambilan keputusan yang lebih baik.

### Analytic Approach:
- Menggunakan algoritma *machine learning* seperti RandomForestClassifier, DecisionTreeClassifier, XGBClassifier, CatBoostClassifier, GradientBoostingClassifier, LGBMClassifier, KNeighborsClassifier, dan PassiveAggressiveClassifier untuk membangun model prediksi churn.
- Melakukan analisis eksploratif pada data pelanggan untuk mengidentifikasi faktor-faktor yang berkontribusi terhadap churn.
- Menggunakan teknik *feature engineering* dan *data preprocessing* untuk memaksimalkan performa model.

### Metric Evaluation:
- **F2-Score (Fokus Utama):** *F2-Score* dipilih sebagai metrik utama karena memberikan penekanan lebih pada recall, dengan tujuan mengidentifikasi sebanyak mungkin pelanggan yang berisiko churn. Ini penting karena biaya untuk mendapatkan pelanggan baru jauh lebih besar dibandingkan dengan biaya yang dikeluarkan akibat salah prediksi (false positive). Dengan mengutamakan recall, perusahaan dapat meminimalkan kehilangan pelanggan yang sebenarnya berisiko churn.
- **F1-Score:** Metrik tambahan untuk memastikan keseimbangan antara precision dan recall. Meskipun F1-Score penting, fokus utama tetap pada F2-Score untuk memaksimalkan penemuan pelanggan berisiko.
- **AUC-ROC:** Menggunakan AUC-ROC sebagai metrik pendukung untuk mengukur kemampuan model dalam membedakan pelanggan churn dan non-churn secara keseluruhan.

Dengan berfokus pada *F2-Score*, WirelessUS dapat memastikan bahwa strategi retensi pelanggan lebih efektif dan tepat sasaran, sehingga mampu mengurangi churn dan mempertahankan pelanggan dengan lebih baik.

## `Univariate` Categorical

```python
cols = ['Dependents', 'OnlineSecurity', 'OnlineBackup', 'InternetService', 'DeviceProtection', 'TechSupport', 'Contract', 'PaperlessBilling']

fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(20, 12), sharey=True)

for i, col in enumerate(cols):
    row = i // 4  
    col_pos = i % 4  
    
    ax = sns.countplot(data=train, x=col, palette='viridis', ax=axes[row, col_pos])
    ax.set_title(f'Proportion of {col}', fontsize=14)
    ax.set_xlabel(' ')
    ax.set_ylabel('Count' if col_pos == 0 else '')
    
    total = len(train)
    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height() / total)
        x = p.get_x() + p.get_width() / 2
        y = p.get_height()
        ax.annotate(percentage, (x, y), ha='center', va='bottom', fontsize=10, color='black', weight='bold')

plt.tight_layout()
plt.show()
```

1. **One-Hot Encoding:**  Mengonversi kolom kategorikal tanpa urutan menjadi kolom biner, sehingga model tidak menganggap nilai numerik sebagai urutan atau tingkatan.

2. **Ordinal Encoding:** Mengonversi kolom dengan urutan (`Contract` dan `tenure_group`) menjadi nilai numerik yang memiliki tingkatan, memungkinkan model untuk menangkap perbedaan tingkat dengan akurat.

3. **RobustScaler:** Digunakan untuk fitur numerik karena data tidak terdistribusi normal.

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, RobustScaler
from sklearn.compose import ColumnTransformer
```

```python
# Step 1: Definisikan kolom untuk one-hot encoding, ordinal encoding, dan kolom numerik
categorical_columns = ['Dependents', 'OnlineSecurity', 'OnlineBackup', 'InternetService', 'DeviceProtection', 'TechSupport', 'PaperlessBilling']
ordinal_columns = ['Contract', 'tenure_group']
numerical_columns = ['TotalCharges', 'MonthlyCharges', 'tenure', 'ratio_monthly_charges']

# Step 2: Definisikan urutan untuk ordinal encoding
contract_order = ['Month-to-month', 'One year', 'Two year']
tenure_group_order = ['New', 'Medium', 'Long']

# Step 3: Buat transformers untuk masing-masing langkah
# OneHotEncoder untuk kolom kategorikal
onehot_transformer = OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore')

# OrdinalEncoder untuk kolom yang memiliki urutan tertentu
ordinal_transformer = OrdinalEncoder(categories=[contract_order, tenure_group_order])

# RobustScaler
scaler = RobustScaler()

# Step 4: Buat ColumnTransformer untuk menggabungkan semua transformasi
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', onehot_transformer, categorical_columns),  # One-hot encoding untuk kolom kategorikal
        ('ord', ordinal_transformer, ordinal_columns),  # Ordinal encoding untuk 'Contract' dan 'tenure_group'
        ('num', scaler, numerical_columns)  # Scaling untuk kolom numerik
    ],
)
```

**Cara kerja ADASYN**:

`ADASYN` (Adaptive Synthetic Sampling Approach) berfungsi menyeimbangkan data dengan menambahkan sampel sintetis di sekitar titik minoritas yang sulit diklasifikasikan. Tidak seperti SMOTE yang membuat sampel sintetis secara merata, ADASYN lebih adaptif karena fokus pada area dengan ketidakseimbangan tinggi. Semakin sulit suatu titik minoritas untuk diklasifikasikan, semakin banyak sampel sintetis yang akan dibuat di sekitarnya. Ini membantu model untuk lebih memahami pola-pola kompleks di kelas minoritas, sehingga meningkatkan akurasi prediksi pada kelas yang kurang terwakili.

**Parameter Tunning :**

1. **`n_estimators`**:
   - *Pengertian*: Ini menentukan berapa banyak pohon yang akan dibangun oleh model. Setiap pohon baru yang ditambahkan belajar dari kesalahan pohon sebelumnya, bertujuan untuk mengurangi error secara bertahap.
   - *Cara Kerja*: Dengan lebih banyak pohon, model akan menjadi lebih kuat dan mampu menangkap pola kompleks. Namun, terlalu banyak pohon bisa membuat model terlalu fokus pada data pelatihan (overfitting), sehingga tidak mampu memprediksi dengan baik pada data baru.

2. **`learning_rate`**:
   - *Pengertian*: Ini mengatur seberapa besar pengaruh setiap pohon terhadap model akhir. Learning rate yang lebih kecil berarti perubahan model terjadi lebih perlahan, sehingga setiap pohon hanya sedikit mengubah hasil akhir.
   - *Cara Kerja*: Jika learning rate rendah, model belajar secara perlahan tapi stabil, mengurangi risiko overfitting karena tidak ada perubahan besar pada setiap iterasi. Namun, ini juga berarti butuh lebih banyak pohon untuk mencapai hasil optimal.

3. **`max_depth`**:
   - *Pengertian*: Ini mengontrol kedalaman maksimum pohon, atau seberapa jauh pohon dapat berkembang untuk membuat keputusan. Semakin dalam pohon, semakin banyak aturan yang dibuat.
   - *Cara Kerja*: Kedalaman yang lebih besar memungkinkan model untuk belajar lebih banyak detail dari data. Tapi jika terlalu dalam, model bisa menjadi terlalu spesifik pada data pelatihan, menangkap noise dan anomali, sehingga kinerjanya menurun pada data baru.

4. **`subsample`**:
   - *Pengertian*: Ini menentukan persentase data pelatihan yang digunakan untuk membangun setiap pohon. Misalnya, subsample 0.8 berarti setiap pohon hanya menggunakan 80% data secara acak.
   - *Cara Kerja*: Dengan tidak menggunakan semua data, model mengurangi korelasi antar pohon. Ini membantu mengurangi overfitting, karena setiap pohon memiliki variasi data yang sedikit berbeda, membuat model lebih kuat terhadap perubahan data.

5. **`min_samples_split`**:
   - *Pengertian*: Ini adalah jumlah minimum sampel yang diperlukan untuk membagi node (cabang) pohon. Jika sebuah node memiliki sampel kurang dari nilai ini, pohon tidak akan membaginya lebih lanjut.
   - *Cara Kerja*: Mengontrol kapan pohon berhenti membagi node. Semakin tinggi nilainya, semakin sedikit pembagian yang terjadi, menghasilkan pohon yang lebih sederhana dan mengurangi risiko model menjadi terlalu rumit dan overfit.

6. **`min_samples_leaf`**:
   - *Pengertian*: Ini adalah jumlah minimum sampel yang harus ada di leaf node (node terakhir). Ini memastikan setiap keputusan akhir dalam pohon didasarkan pada jumlah data yang cukup.
   - *Cara Kerja*: Mencegah leaf node yang hanya berisi sedikit data, yang bisa membuat model rentan terhadap noise. Dengan memastikan leaf node memiliki cukup sampel, model menjadi lebih stabil dan kurang cenderung menangkap pola yang tidak relevan atau anomali pada data pelatihan.

Setiap parameter dalam tuning ini berfungsi untuk mengontrol bagaimana model belajar dari data. Tuning ini mengatur jumlah pohon, durasi belajar, kedalaman pohon, serta batasan untuk pembagian node, memastikan model tidak terlalu underfit atau terlalu kompleks (overfit). Ini membuat model dapat memahami pola yang ada pada data. 

Saya juga disini mempertimbangkan durasi pelatihan model sangat penting, karena ketika volume data meningkat, waktu pelatihan akan bertambah secara eksponensial. Jika waktu pelatihan pada data saat ini sudah lama, maka penambahan data di masa depan bisa menyebabkan durasi pelatihan meningkat drastis, menghambat proses iterasi dan pengembangan model. Oleh karena itu, diperlukan parameter tuning dan optimasi yang efisien untuk memastikan model tetap scalable dan dapat diproses dalam waktu yang wajar meskipun data semakin besar.

```python
pipeline = ImbPipeline(steps=[
    ('preprocessor', preprocessor),  
    ('resampler', ADASYN(random_state=99)),  
    ('feature_selection', SelectKBest(score_func=f_classif)),
    ('classifier', GradientBoostingClassifier(random_state=99))
])


param_grid = {
    'classifier__n_estimators': [50, 100, 150],  # Jumlah estimasi pohon
    'classifier__learning_rate': [0.01, 0.05, 0.1],  # Rentang learning rate
    'classifier__max_depth': [3, 4, 5],  # Kedalaman pohon 
    'classifier__subsample': [0.8, 1.0],  # Proporsi sampel data yang digunakan untuk membangun setiap pohon
    'classifier__min_samples_split': [2, 5, 10],  # Minimum jumlah sampel untuk membagi node
    'classifier__min_samples_leaf': [1, 2, 4]  # Minimum jumlah sampel di leaf node
}


grid_search = GridSearchCV(
    pipeline, 
    param_grid=param_grid, 
    scoring={'f1': 'f1', 'f2': f2_scorer}, 
    refit='f2',  
    cv=5,  
    n_jobs=-1
)

start_time = time.time()
grid_search.fit(X_train, y_train)
end_time = time.time()

fitting_time = end_time - start_time
print(f"Fitting time: {fitting_time:.2f} seconds")

cv_results = grid_search.cv_results_

results_df_ada_hyperparam5 = pd.DataFrame({
    'Model': ['GradientBoost Classifier'],
    'Mean F1 Score': [cv_results['mean_test_f1'][grid_search.best_index_]],
    'Std F1 Score': [cv_results['std_test_f1'][grid_search.best_index_]],
    'Mean F2 Score': [cv_results['mean_test_f2'][grid_search.best_index_]],
    'Std F2 Score': [cv_results['std_test_f2'][grid_search.best_index_]]
})

print("GSCV Results - GradientBoost Hyperparameter Tuning:")
display(results_df_ada_hyperparam5)


best_model = grid_search.best_estimator_

y_train_pred = best_model.predict(X_train)
y_test_pred = best_model.predict(X_test)

print("Classification Report - Train Set")
train_report = classification_report(y_train, y_train_pred, output_dict=True)
display(pd.DataFrame(train_report).transpose())

print("Classification Report - Test Set")
test_report = classification_report(y_test, y_test_pred, output_dict=True)
display(pd.DataFrame(test_report).transpose())
```

### Pembuatan Rule-Based Model untuk Prediksi Churn dari perusahaan WirelessUS

1. **Identifikasi Faktor Risiko:**
   - Tentukan variabel `Contract`, `Tenure`, `MonthlyCharges`, dan penggunaan `Layanan Tambahan`.
   - Gunakan data historis untuk menentukan pola perilaku yang terkait dengan churn.

2. **Pembuatan Aturan Dasar:**
   - **Contract Month-to-Month:** Skor +1. Pelanggan dengan kontrak bulanan memiliki fleksibilitas untuk berhenti kapan saja.
   - **Tenure Rendah (< 6 Bulan):** Skor +2. Pelanggan baru lebih mungkin churn karena belum merasa terikat.
   - **Biaya Bulanan Tinggi (> $50):** Skor +2. Biaya tinggi bisa menyebabkan ketidakpuasan.
   - **Tidak Menggunakan Layanan Tambahan:** Skor +1 per layanan. Ini menunjukkan kurangnya manfaat yang dirasakan pelanggan.

3. **Kombinasi Aturan:**
   - Berikan skor tambahan untuk kombinasi faktor risiko, pelanggan dengan tenure rendah dan kontrak bulanan mendapatkan skor ekstra +3.

4. **Penetapan Threshold:**
   - Tentukan ambang batas total skor ≥ 5, untuk mengidentifikasi pelanggan yang berisiko churn.

- Rulebase ini didasarkan pada pendekatan yang dijelaskan oleh Fader dan Hardie dalam analisis basis pelanggan yang menggunakan data perilaku untuk memprediksi churn. [Source](https://faculty.wharton.upenn.edu/wp-content/uploads/2012/04/Fader_hardie_jim_09.pdf).

```python
def combined_rule_based_prediction(row):
    score = 0
    
    # Aturan 1: Tidak menggunakan layanan tambahan
    no_additional_services = ['No', 'No internet service']
    if row['OnlineSecurity'] in no_additional_services:
        score += 1
    if row['TechSupport'] in no_additional_services:
        score += 1
    if row['OnlineBackup'] in no_additional_services:
        score += 1
    if row['DeviceProtection'] in no_additional_services:
        score += 1
    
    # Aturan 2: Contract month-to-month - Skor +1
    if row['Contract'] == 'Month-to-month':
        score += 1
    
    # Aturan 3: Tenure sangat rendah (< 6 bulan) dan biaya bulanan tinggi - Skor +2
    if row['tenure'] < 6 and row['MonthlyCharges'] > 50:
        score += 2
    
    # Aturan 4: Kombinasi Tenure rendah dan Contract month-to-month - Skor ekstra +3
    if row['tenure'] < 6 and row['Contract'] == 'Month-to-month':
        score += 3
    
    # Threshold: jika total score >= 5 maka churn
    if score >= 5:
        return 1
    else:
        return 0

test_rule_based = test.copy()
test_rule_based['Churn_Predicted_Scoring'] = test_rule_based.apply(combined_rule_based_prediction, axis=1)


test_report = classification_report(test_rule_based['Churn'], test_rule_based['Churn_Predicted_Scoring'], output_dict=True)
print("Classification Report - Test Set with Rule-Based")
display(pd.DataFrame(test_report).transpose())

print("F2 score (Test):", fbeta_score(test_rule_based['Churn'], test_rule_based['Churn_Predicted_Scoring'], beta=2))
```

## **Perbandingan dan Pengurangan Kerugian Tahunan:**
#### **Skenario:**
- **Biaya Akuisisi Pelanggan Baru (CAC):** $200 per pelanggan.
- **Biaya Retensi Pelanggan (Retention Cost):** $50 per pelanggan.
- **Jumlah Total Pelanggan yang Berisiko Churn:** 500 pelanggan.
- **Periode Analisis:** 12 bulan (1 tahun).

#### **Perhitungan untuk Rule-Based:**
1. **False Negative (FN):** 146 pelanggan.
   - Kerugian dari FN = 146 * $200 (CAC) = $29,200.
2. **False Positive (FP):** 190 pelanggan.
   - Biaya FP = 190 * $50 (Retention Cost) = $9,500.
3. **Total Kerugian Bulanan:**
   - Kerugian = Kerugian FN + Biaya FP = $29,200 + $9,500 = $38,700.
4. **Total Kerugian Tahunan:**
   - Kerugian Tahunan = $38,700 * 12 = $464,400.

#### **Perhitungan untuk Machine Learning:**
1. **False Negative (FN):** 31 pelanggan.
   - Kerugian dari FN = 31 * $200 (CAC) = $6,200.
2. **False Positive (FP):** 364 pelanggan.
   - Biaya FP = 364 * $50 (Retention Cost) = $18,200.
3. **Total Kerugian Bulanan:**
   - Kerugian = Kerugian FN + Biaya FP = $6,200 + $18,200 = $24,400.
4. **Total Kerugian Tahunan:**
   - Kerugian Tahunan = $24,400 * 12 = $292,800.

<hr>

1. **Total Pengurangan Kerugian Tahunan:**
    $$ \text{Pengurangan Kerugian} = \text{Kerugian Tahunan}_{Rule-Based} - \text{Kerugian Tahunan}_{ML} $$
    $$ \text{Pengurangan Kerugian} = 464,400 - 292,800 = 171,600 \text{ USD} $$

2. **Persentase Pengurangan Kerugian:**
    $$ \text{Persentase Pengurangan} = \frac{\text{Kerugian Tahunan}_{Rule-Based} - \text{Kerugian Tahunan}_{ML}}{\text{Kerugian Tahunan}_{Rule-Based}} \times 100 $$
    $$ \text{Persentase Pengurangan} = \frac{464,400 - 292,800}{464,400} \times 100 \approx 36.96\% $$


Skenario ini dibuat apabila data selalu diolah tiap bulan dan diakumulasikan dalam 12 bulan (1 tahun)
